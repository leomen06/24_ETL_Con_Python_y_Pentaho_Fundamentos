# <a name="mark_01"></a>Algunos Conceptos Generales: 
### [Index](#index)

Conceptos

* Source (fuente): 
De dónde voy a sacar los datos, estructuradas, no estructuradas, APIs, CRM.

* Target (Objetivo): 
donde voy a cargar el proceso de ETL, por lo general base de datos unificada columnar (snowflake, redshift, big query) o transaccionales (Oracle, postgreSQL, mysql).

* Staging: 
Area de paso (suele ser en memoria), dónde se van a hacer transformaciones. Se pueden guardar en bases de datos para luego retomar la transformación y proceder al load.

* Data Warehouse: 
Bodega de datos, almacenar datos de una manera definida.

* [Star model](https://towardsdatascience.com/star-schema-924b995a9bdf): 
Una manera de cargar todas las fuentes y estructuras al Data Warehouse. una tabla con solo métricas importante, y tablas alrededor llamadas dimensiones (se conectan con las metrices usualmente con Foreign key).

“El modelo estrella se compone de una tabla central llamada tabla de hechos, que contiene información cuantitativa y numérica (como ventas, ingresos, cantidad de productos vendidos, etc.) y varias tablas dimensionales que contienen información descriptiva y contextual (como fecha, hora, ubicación, productos, clientes, etc.).”

* Data Lake: 
Repositorio dónde se manda información, formato nativo o raw de los datos (datos sin modificaciones, cómo los genera el origen). Se guarda en disco duro.

* Data Lakehouse (Delta-Lake):
Data Warehouse + Data Lake: se guarda de manera raw de los datos, pero se tiene una capa de consulta (sql, spark) para realizar analítica sin mover los datos. Se una con el proceso ELT.

````python
import pandas as pd
from sqlalchemy import create_engine
````

````python
#engine=create_engine('postgresql+psycopg2://postgres:Postgresql_3967@localhost/postgres')
engine=create_engine('postgresql+psycopg2://postgres:mysecretpass@localhost/postgres')
````

Creando una instancia de la clase `Engine` en SQLAlchemy para conectarse a una base de datos PostgreSQL utilizando el dialecto `psycopg2`.

Aquí está la explicación paso a paso:

1. `create_engine`: Esto es una función proporcionada por SQLAlchemy que se utiliza para crear una instancia de la clase `Engine`.

2. `'postgresql+psycopg2://`: Esto define el dialecto y el controlador que se utilizará para conectarse a la base de datos PostgreSQL. `'postgresql'` es el dialecto que representa a PostgreSQL, y `'psycopg2'` es el controlador de base de datos que se utilizará para comunicarse con PostgreSQL.

3. `postgres:mysecretpass@localhost/postgres'`: Esta parte es la cadena de conexión, que contiene la información necesaria para conectarse a la base de datos. Aquí están los detalles:

   - `postgres`: Este es el nombre de usuario de la base de datos al que se conectará.
   - `mysecretpass`: Esta es la contraseña del usuario.
   - `localhost`: Esta es la dirección del servidor de la base de datos. En este caso, está configurado para el mismo sistema en el que se está ejecutando el código, lo que indica que la base de datos se encuentra en la misma máquina.
   - `postgres`: Este es el nombre de la base de datos a la que se conectará.

En resumen, esta sentencia crea una instancia de la clase `Engine` que se utilizará para conectarse a una base de datos PostgreSQL utilizando el dialecto `psycopg2`. Se proporciona la información necesaria en la cadena de conexión para autenticarse en el servidor de la base de datos y acceder a la base de datos específica.

# <a name="mark_06"></a>Transformación: 
### [Index](#index)

````python
#Quitando valores nulos de 'Code_comm'.
df_codes=df_codes[df_codes['Code_comm'].notnull()]
df_codes
````

````python
#Esta función extrae los primeros 5 o 6 valores de los códigos de 11 o más dígitos
def clean_code(text):
    
    text=str(text)
    parent_code=None
    
    if len(text)==11:
        code=text[:5]
        parent_code=text[:1]
    else: 
        code=text[:6]
        parent_code=text[:2]
    try:
        parent=df_parents[df_parents['Code_comm']==parent_code]['Description'].values[0]
    except:
        parent="There is not description to show"
    return (code, parent)
````

````python
#Aplicando la función a cada uno de los rows, crea 2 nuevas columnas 'clean_code', 'parent_description'.
#con valores filtrados por la función "clean_code"
df_codes[['clean_code', 'parent_description']]=df_codes.apply(lambda x : clean_code(x['Code']), axis=1, result_type='expand')
df_codes
````

````python
#limpiando de valores nulos.
df_codes=df_codes[df_codes['clean_code'].notnull()]
df_codes
````

````python
#tomando solo 3 columnas de todo el df.
df_codes=df_codes[['clean_code', 'Description','parent_description']]
df_codes
````

````python
#Agregando "id_code":
df_codes['id_code']=df_codes.index
df_codes
````

````python
#parcear "clean_code" de str a int.
df_codes['clean_code']=df_codes['clean_code'].astype('int64')
df_codes
````

````python
#Función que vuelve valores de columnas en un data frame.
def create_dimension(data, id_name):#recibe una serie y un id
    list_keys=[]
    value=1
    for _ in data:#itera lo que llegue de parámetro en los datos
        list_keys.append(value)
        value=value+1
    return pd.DataFrame({id_name:list_keys, 'values':data})
    
````

En Python, la expresión `for _ in data:` se utiliza para iterar sobre los elementos de una secuencia, como una lista, una tupla, un diccionario u otro tipo de iterable. Sin embargo, el uso del guion bajo (`_`) como nombre de variable es una convención que indica que no estás interesado en el valor real de cada elemento durante la iteración.

En otras palabras, cuando usas `for _ in data:`, estás diciendo que solo estás interesado en la repetición de la iteración y no necesitas el valor real de cada elemento. Esto es común cuando solo estás interesado en realizar una acción repetida cierta cantidad de veces o cuando solo necesitas recorrer una secuencia sin utilizar los valores individuales.

Por ejemplo:

```python
numbers = [1, 2, 3, 4, 5]

for _ in numbers:
    print("Iterating!")

# Output:
# Iterating!
# Iterating!
# Iterating!
# Iterating!
# Iterating!
```

En este caso, el guion bajo se utiliza para indicar que no necesitamos el valor real de los elementos en la lista `numbers`, simplemente estamos iterando cinco veces e imprimiendo "Iterating!" en cada iteración.

En resumen, `for _ in data:` es una forma de iterar sobre una secuencia cuando no te importa el valor de los elementos, sino solo la acción repetida.

````python
import os
import boto3 #conección a "S3"
import redshift_connector as rc

client=boto3.client(
    's3',#tipo de conección
    #Debo tener creadas mis variables de entorno "AWS_ACCESS_KEY_ID", y "AWS_SECRET_ACCESS_KEY"
    aws_access_key_id=os.environ.get('AWS_ACCESS_KEY_ID'),
    aws_secret_access_key=os.environ.get('AWS_SECRET_ACCESS_KEY')
)

#nuevamente todos los datos se encuentran en la variable de entorno.
conn=rc.connect(
    host=os.environ.get('redshift_host'),
    database=os.environ.get('redshift_database'),
    port=5439,
    user=os.environ.get('redshift_user'),
    password=os.environ.get('redshift_pass')
)

#Abriendo conección en redshift.
cursor=conn.cursor()
````

````python
#Subri archivos a S3 y hacer un copy a redshift para cargar a travéz de S3.
def load_file(file_name):
    table_name=file_name.split('.')[0]# ej "file.csv", solo toma "file"
    client.upload_file(
        Filename=f'target/{file_name}',
        Bucket='platzi-etl',#Bucket en S3
        key=f'course_etl_target{file_name}/'#Cómo se cargará el archivo
    )
    #todos los parámetros para la conección y carga, hay que revisar 
    sentence=''f'copy etl_test.{table_name}' from f's3://platzi_etl_target/{file_name}' credentials f'aws_access_key_id={os.environ.get('AWS_ACCESS_KEY_ID')}';
    f'aws_secret_access_key={os.environ.get('AWS_SECRET_ACCESS_KEY')}' csv delimiter '|' region 'us-west-2' ignoreheader 1 ''
    try:
        cursor.execute(sentence)
        print('Ok en la tabla '+table_name)
    except:
        print('Error en la tabla '+table_name)
````

### [Index](#index)

En la clase anterior definimos esta función de carga de archivos en nuestra notebook de Python:
![](https://static.platzi.com/media/user_upload/sentence-85144aa1-bb6c-4cb3-bdde-27e2b85db1f6.jpg)

💡Dentro de esta función esta la variable sentence, que es una secuencia de texto con una instrucción para copiar los registros de los archivos CSV, que están dentro del bucket de S3, hacia la base de datos dev que está en un schema.

Esta sentencia debe configurarse específicamente para el schema de tu base de datos y la dirección de tu bucket en S3.

Revisitemos tres componentes principales dentro de la sentencia:
![](https://static.platzi.com/media/user_upload/sentence_partes-e47a62d8-fdc1-40b1-bc1f-d1e358ae8531.jpg)

- Schema en DB en Redshift: es el schema de nuestra base de datos donde están las tablas a donde se copiarán los registros de los archivos CSV. Este debe ser remplazado por ese schema, que en tu caso debería ser public en lugar de etl_test

- Nombre de bucket de S3: dirección del bucket donde se guardan temporalmente los archivos CSV. En tu caso debe remplazarse 's3://platzi-etl/course_etl_target/{}' por el nombre único de tu bucket. Si tu bucket se llamara nombre-unico-bucket-curso-platzi en la sentencia debería estar 's3://nombre-unico-bucket-curso-platzi{}'

- Región elegida de AWS: es esa región donde creaste el clúster de Redshift. En caso de que no lo hayas hecho en la región us-west-2 escribe el código de la región que elegiste.

Tu sentence debería verse de forma similar a la siguiente, pero recuerda que no será idéntica porque el nombre de tu bucket es único en el mundo:

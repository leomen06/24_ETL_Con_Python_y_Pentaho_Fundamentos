Conceptos

* Source (fuente): 
De d贸nde voy a sacar los datos, estructuradas, no estructuradas, APIs, CRM.

* Target (Objetivo): 
donde voy a cargar el proceso de ETL, por lo general base de datos unificada columnar (snowflake, redshift, big query) o transaccionales (Oracle, postgreSQL, mysql).

* Staging: 
Area de paso (suele ser en memoria), d贸nde se van a hacer transformaciones. Se pueden guardar en bases de datos para luego retomar la transformaci贸n y proceder al load.

* Data Warehouse: 
Bodega de datos, almacenar datos de una manera definida.

* [Star model](https://towardsdatascience.com/star-schema-924b995a9bdf): 
Una manera de cargar todas las fuentes y estructuras al Data Warehouse. una tabla con solo m茅tricas importante, y tablas alrededor llamadas dimensiones (se conectan con las metrices usualmente con Foreign key).

El modelo estrella se compone de una tabla central llamada tabla de hechos, que contiene informaci贸n cuantitativa y num茅rica (como ventas, ingresos, cantidad de productos vendidos, etc.) y varias tablas dimensionales que contienen informaci贸n descriptiva y contextual (como fecha, hora, ubicaci贸n, productos, clientes, etc.).

* Data Lake: 
Repositorio d贸nde se manda informaci贸n, formato nativo o raw de los datos (datos sin modificaciones, c贸mo los genera el origen). Se guarda en disco duro.

* Data Lakehouse (Delta-Lake):
Data Warehouse + Data Lake: se guarda de manera raw de los datos, pero se tiene una capa de consulta (sql, spark) para realizar anal铆tica sin mover los datos. Se una con el proceso ELT.

````python
#Esta funci贸n extrae los primeros 5 o 6 valores de los c贸digos de 11 o m谩s d铆gitos
def clean_code(text):
    
    text=str(text)
    parent_code=None
    
    if len(text)==11:
        code=text[:5]
        parent_code=text[:1]
    else: 
        code=text[:6]
        parent_code=text[:2]
    try:
        parent=df_parents[df_parents['Code_comm']==parent_code]['Description'].values[0]
    except:
        parent="There is not description to show"
    return (code, parent)
````

````python
#Funci贸n que vuelve valores de columnas en un data frame.
def create_dimension(data, id_name):#recibe una serie y un id
    list_keys=[]
    value=1
    for _ in data:#itera lo que llegue de par谩metro en los datos
        list_keys.append(value)
        value=value+1
    return pd.DataFrame({id_name:list_keys, 'values':data})
    
````

En Python, la expresi贸n `for _ in data:` se utiliza para iterar sobre los elementos de una secuencia, como una lista, una tupla, un diccionario u otro tipo de iterable. Sin embargo, el uso del guion bajo (`_`) como nombre de variable es una convenci贸n que indica que no est谩s interesado en el valor real de cada elemento durante la iteraci贸n.

En otras palabras, cuando usas `for _ in data:`, est谩s diciendo que solo est谩s interesado en la repetici贸n de la iteraci贸n y no necesitas el valor real de cada elemento. Esto es com煤n cuando solo est谩s interesado en realizar una acci贸n repetida cierta cantidad de veces o cuando solo necesitas recorrer una secuencia sin utilizar los valores individuales.

Por ejemplo:

```python
numbers = [1, 2, 3, 4, 5]

for _ in numbers:
    print("Iterating!")

# Output:
# Iterating!
# Iterating!
# Iterating!
# Iterating!
# Iterating!
```

En este caso, el guion bajo se utiliza para indicar que no necesitamos el valor real de los elementos en la lista `numbers`, simplemente estamos iterando cinco veces e imprimiendo "Iterating!" en cada iteraci贸n.

En resumen, `for _ in data:` es una forma de iterar sobre una secuencia cuando no te importa el valor de los elementos, sino solo la acci贸n repetida.
